```python
df = pd.read_excel('data/thresh24.xlsx', skiprows=5)
```

```python
df=df.rename(columns={"Unnamed: 0": "household","Unnamed: 1": "weighted_average"})
```

```python
df=df[~df.household.str.contains('65', na=False)]
```

```python
df
```

```python
df_melt=pd.melt(df, id_vars=['household'],var_name='children', value_name='number')
df_melt=df_melt[~df_melt.number.isna()]
```

```python
df_melt['threshold'] = np.floor(df_melt.number / 5000.0) * 5000.0
```

```python
df_melt_12=df_melt[df_melt.household.isin(['One person (unrelated individual):','Two people:'])]
```

```python
df_melt_12['household']=[1,2]
df_melt_12=pd.concat([df_melt_12,df_melt_12.loc[[3]]],ignore_index=True)
df_melt_12['children']=[0,0,1]
```

```python
df_melt_rest=df_melt[~df_melt.household.isin(['One person (unrelated individual):','Two people:'])]
df_melt_rest=df_melt_rest[df_melt_rest.children !="weighted_average"]
```

```python
household_conv={'Three people':3,'Four people':4,'Five people':5,'Six people':6,'Seven people':7,'Eight people':8,'Nine people or more':9}
child_conv={' None':0,'One':1,'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight or more':8}
df_melt_rest['household']=[household_conv[i] for i in df_melt_rest['household']]
df_melt_rest['children']=[child_conv[i] for i in df_melt_rest['children']]
```

```python
df_melt2=pd.concat([df_melt_12,df_melt_rest],ignore_index=True)
df_melt2=df_melt2.drop('number', axis=1)
```

```python
df_melt2.to_csv('data/thresh24.csv', index=False)
```
